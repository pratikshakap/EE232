```r
print("For m=2")
g1<-barabasi.game(1000, m = 2,directed=FALSE)
fg1 <- fastgreedy.community(g1)
mod <- modularity(fg1)
cat("Modularity for m=2 and n=1000",mod)
print(" ")
g2<-barabasi.game(10000, m = 2,directed=FALSE)
fg2 <- fastgreedy.community(g2)
mod <- modularity(fg2)
cat("Modularity for m=2 and n=10000",mod)

plot(fg1,g1,vertex.size=5, vertex.label.cex=0.3,main="m=2 and n=1000")
plot(fg2,g2,vertex.size=5, vertex.label.cex=0.3,main="m=2 and n=10000")
```

```r
g1<-barabasi.game(1000, m = 2,directed=FALSE)


g2<-barabasi.game(10000, m = 2,directed=FALSE)
plot(x=degree_distribution(g1),log="xy",type="l",col="red",main="Degree distribution on a log-log scale for m=2",xlab=" ",ylab="")
lines(x=degree_distribution(g2),log="xy",type="l",col="blue")

legend("topright",c("n=1000","n=10000"),col=c("red","blue"),lty=1, cex=0.8)
```

```r
degrees<-list()
i<-1
for (n in sample(vcount(g1), 1000)){
    if(length(neighbors(g1,n))==1){
        n1<-neighbors(g1,n)
    }
    else {
        n1<-sample(neighbors(g1,n),1)
    }
    degrees[[i]]<-degree(g1,n1)
    i<-i+1
}
deg<-table(unlist(degrees))
deg<-deg/sum(table(unlist(degrees)))
plot(x=deg,type="l",col="red",log="x",main="Degree distribution of random nodes for m=2",ylab="")
plot(x=degree_distribution(g1),type="l",col="blue",main="Normal vs random ",ylab="",xlab="")
lines(x=deg,type="l",col="red")
legend("topright",c("Normal","Random"),col=c("blue","red"),lty=1, cex=0.8)
```

```r
g<-barabasi.game(1000, m = 2,directed=FALSE)
degrees<-list()
i<-1
for(n in seq(1,1000,1)){
    degrees[[i]]<-degree(g,n)
    i<-i+1
}
deg<-unlist(degrees)

plot(x=deg,col="red",main="Age vs degree for m=2",xlab="Age ",ylab="Degree")
```
