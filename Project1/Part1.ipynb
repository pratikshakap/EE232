```r
#1.1.a
for(p in c( 0.003, 0.004, 0.01, 0.05,0.1)){
    g <- erdos.renyi.game(1000, p)
    hist(degree(g))
    cat ("For probability ",p, ": Mean is ",mean(degree(g))," and variance is ",var(degree(g)))
}
```

```r
#1.1.b
#To numerically estimate the probability that the graphs are connected:
for(p in seq(0, 1, 0.01))
    {
    g<-erdos.renyi.game(1000, p, directed=F)
    if(!is.connected(g))
        {
        g.components <- clusters(g)
        ix <- which.max(g.components$csize)
        g.giant <- induced.subgraph(g, which(g.components$membership == ix))
        #print(g.components$csize)
        dm=diameter(g.giant, directed = FALSE, unconnected = TRUE, weights = NULL)
        cat(p,"     diamater",dm, "    size",vcount(g.giant),"\n")
    }
    else 
        cat(p, "  CONNECTED \n")
}
```

```r
#1.1.c
mylist<-list()
val<-list()
i<-1
for (p in seq(0.000, 0.01, by = 0.0002)){
    for(j in 1:101){
        g <- erdos.renyi.game(1000, p)    
        g.components <- clusters(g)
        ix <- which.max(g.components$csize)
        g.giant <- induced.subgraph(g, which(g.components$membership == ix))
        #print(paste(p," ",vcount(g.giant)))
        mylist[[i]]<-p
        val[[i]]<-vcount(g.giant)/1000
        i<-i+1
    }
    
    
}
plot(x=mylist,y=val,col="blue",main="Normalized GCC vs p",xlab="p",ylab="Normalized GCC")
```

```r
#1.1.d i
avg=0.5
sizelist05<-list()
nlist<-list()
i<-1
for (n in seq(100, 10000, by = 100)){
    g <- erdos.renyi.game(n, avg/n)    
    g.components <- clusters(g)
    ix <- which.max(g.components$csize)
    g.giant <- induced.subgraph(g, which(g.components$membership == ix))   
    sizelist05[[i]]<-vcount(g.giant)
    nlist[[i]]<-n
    i<-i+1
}
plot(x=nlist,y=sizelist05,col="red",main="Average degree of node=0.5",xlab="Number of nodes",ylab="Size of GCC")
```

```r
#1.2.e
degrees<-list()
i<-1
for (n in sample(vcount(g1), 100)){
    if(length(neighbors(g1,n))==1){
        n1<-neighbors(g1,n)
    }
    else {
        n1<-sample(neighbors(g1,n),1)
    }
    degrees[[i]]<-degree(g1,n1)
    i<-i+1
}
deg<-table(unlist(degrees))
deg<-deg/sum(table(unlist(degrees)))
plot(x=deg,type="l",col="red",log="x")
```
